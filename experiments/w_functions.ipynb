```python
import sys
import os

SCRIPT_DIR = os.path.dirname(os.path.abspath(os.getcwd()))
sys.path.append(SCRIPT_DIR)

import numpy as np

from tueplots import bundles

from rcgp.w import IMQ, SE

import matplotlib.pyplot as plt
CB_color_cycle = ['#377eb8', '#ff7f00', '#4daf4a',
                  '#f781bf', '#a65628', '#984ea3',
                  '#999999', '#e41a1c', '#dede00']
```

```python
plt.rcParams['figure.figsize'] = (12, 5)
```

```python
y = np.linspace(-5, 5, 500)
x = np.linspace(-5, 5, 500)
```

```python
w1 = IMQ()
w2 = SE()
```

```python
with plt.rc_context(bundles.aistats2023()):
    fig, ax = plt.subplots(figsize = (3.25, 1.5), dpi = 200)
    ax.plot(y, w1.W(x,y), label = 'IMQ', c = CB_color_cycle[0], lw=2)
    ax.plot(y, w2.W(x,y), label = 'SE', c = CB_color_cycle[1], lw=2)
    ax.vlines([-1,1],ymin=0,ymax=1, label= r'c = 1', ls= '--', colors=CB_color_cycle[4], lw=2)
    ax.hlines(1,xmin=-5,xmax=5, label = r'$\beta = 1$', ls= '--', colors=CB_color_cycle[6], lw=2)
    ax.set_xlabel(r'$y-m(x)$')
    ax.set_ylabel(r'$w(x,y)$')
    plt.legend()
    fig.savefig('figures/w.pdf', format="pdf", bbox_inches= "tight")
    plt.show()
```
